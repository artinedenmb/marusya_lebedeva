```python
def func1(arr):
    array = np.array([(1 / np.exp(np.sin(arr) + 1)) / (5 / 4 + 5 * arr)])  
    base = 1 + np.sqrt(arr)
    exponent = np.emath.logn(base, array)
    return exponent
```

```python
print(func1(np.array([1, 10, 1000])))
```

упражнение 3

```python
def abracadabra(x):
    mainthg = (x*x + 1) * np.exp((-1) * np.abs(x) / 10)
    osne = 1 + np.tan(1 / (1 + np.sin(x)**2))
    lthm = np.log(mainthg) / np.log(osne)
    return lthm
```

```python
plt.figure(figsize=(20, 15))
x = np.linspace(-20, 20)
plt.plot(x, abracadabra(x))
```

```python
fig, ax = plt.subplots(figsize = (20, 15))
x = [1, 2, 3, 4, 5, 6]
y = [1, 1.42, 1.76, 2, 2.24, 2.5]
ax.set_xlim(0, 7)
ax.set_ylim(0, 3)

p, v = np.polyfit(x, y, deg=1, cov=True)
pp = np.poly1d(p)
ax.plot(x, pp(x), 'o-')

p2, v2 = np.polyfit(x, y, deg=2, cov=True)
pp2 = np.poly1d(p2)
ax.plot(x, pp2(x), '*-')

ax.scatter(x, y, color='purple')


```
